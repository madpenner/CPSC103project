````python
@typecheck
def is_bigger(f1: Footprint, f2: Footprint)-> bool:
    """
    takes two footprints and returns True if the first amount is bigger than the second
    """
    #return True #stub
    #template from two Footprints
    if f1.amt == None:
        return False
    elif f2.amt == None:
        return True
    else:
        return f1.amt > f2.amt

@typecheck
def same_amt(f1: Footprint, f2: Footprint)-> bool:
    """
    takes two footprints and returns True if they are the same amount
    """
    #return True #stub
    #template from two Footprints
    return f1.amt == f2.amt

@typecheck
def same_name(f1: Footprint, f2: Footprint)-> bool:
    """
    takes two footprints and returns True if they are the same amount
    """
    #return True #stub
    #template from two Footprints
    return f1.name == f2.name

@typecheck
def is_biggest(lof: List[Footprint])-> List[Footprint]:
    """
    takes a list of Footprints and returns a list of Footprints containing 
    the Footprint(s) with the biggest amount. 
    Returns two Footprints if the two biggest amounts are tied
    """
    #return [] #stub
    #template from List[Footprint]
    
    #start 'biggest' with the smallest value so that any value recorded for a Footprint will be bigger and 
    #will be returned instead. 
    
    #If there are no values recorded for any of the footprints in the list of footprints for a country, 
    #that countries biggest footprint will be recorded as Footprint('none', 0) to 
    #represent that there is no footprint data for that country.
    
    biggest = Footprint('none', 0)
    
    if lof == []:
        return []
    
    for f in lof:
        if is_bigger(f, biggest):
            biggest = f
            
    for f in lof: 
        if (same_amt(biggest, f)) and (same_name(biggest, f) == False):
            return [biggest, f]
        
    return [biggest]

@typecheck
def biggest_footprint(c: Country)-> List[Footprint]:
    """
    takes a Country and returns the Footprint(s) with the biggest amount. 
    Returns two Footprints if the two biggest amounts are tied
    """
    #return [] #stub
    #template from Country
    
    return is_biggest(c.footprints)

@typecheck
def all_footprints(loc: List[Country])-> List[Footprint]:
    """
    takes a list of countries and returns a list of Footprints representing all the Footprints that are
    the largest Footprint(s) for each country in the list
    """ 
    #return [] #stub
    #template based on list[Country]
    
    footprints = [] #Type: List[Footprint] 
    for c in loc:
        footprints = footprints + biggest_footprint(c) 
    
    return footprints

@typecheck
def count_names(lof: List[Footprint], footprint: str)-> int:
    """
    takes a list of footprints and a footprint name and returns the number of the amount of times 
    a footprint has the same name as the given footprint name
    """
    #return 0 #stub
    #template from List[Footprint]
    
    #count is the amount of Footprints seen so far that have the same name as the given footprint name 
    count = 0 #Type: int
    for f in lof:
        if f.name == footprint:
            count = count + 1
    
    return count 

@typecheck
def show_labels(loc: List[Country], los: List[str]) -> List[str]:
    """
    takes a list of strings representing the labels for the footprints in the list of countries given.
    returns only the ones where there are values present for them from the list of countries. 
   
    eg. if the list of countries countains a value for every name in the list, every name will be returned.
    if there is only a value for the name "carbon", only carbon will be returned in the list of strings.
    """
    #return [] #stub
    #template based on List[str] with additional parameter List[Country]
    
    Cropland = "Cropland"
    Grazing = "Grazing"
    Forest = "Forest"
    Carbon = "Carbon"
    Fish = "Fish"
    
    for s in los:
        if count_names(all_footprints(loc), 'cropland')== 0:
            Cropland = " "
        if count_names(all_footprints(loc), 'grazing')== 0:
            Grazing = " "
        if count_names(all_footprints(loc), 'forest')== 0:
            Forest = " "
        if count_names(all_footprints(loc), 'carbon')== 0:
            Carbon = " "
        if count_names(all_footprints(loc), 'fish')== 0:
            Fish = " "
    return [Cropland, Grazing, Forest, Carbon, Fish]

@typecheck
def show_pie_chart(loc: List[Country]) -> None: # Analyze renamed to be show_pie_chart
    """
    show a pie chart displaying the frequency of each footprint as the biggest footprint for each 
    country in loc. Every time a certain footprint is the biggest for a country, its count goes up, and that 
    count is charted in the pie graph to see which factor has the highest count of being the biggest footprint
    for a country. ie. If the first country in loc has a list of footprints where the Carbon footprint is the 
    largest, the count for carbon becomes 1 while the other counts remain at 0. 
    
    Labels and Names must have the same length. If there are no values for any of Cropland, Grazing, 
    Forest, Carbon, or Fish, (ie. count_names is 0), then the corresponding label for that wedge will
    not be shown on the graph. 
    
    The percentages of everything not labeled (or taking up space in the pie) will add up to 0%
    
    "explode" (set out from the circle) each piece of the pie as specified in explode.
    
    returns None
    """
    #return None  #stub
    # Template based on visualization
    
    # Plan:
    # 1) Find the count for all the footprints in loc for "cropland" (how many times cropland appears in the 
    #    list)
    # 2) repeat step one for "grazing", "forest", "carbon", and "fish"
    # 3) find the labels based on the counts for each footprint (no label for a footprint if count is 0)
    # 4) plot the counts from steps 1 and 2 with the labels from step 3 in a pie chart containing 0 for 
    #    explode and the counts in %'s relative to the other counts. (% of the chart that it takes up)

    
    Cropland = count_names(all_footprints(loc), 'cropland')
    Grazing = count_names(all_footprints(loc), 'grazing')
    Forest = count_names(all_footprints(loc), 'forest')
    Carbon = count_names(all_footprints(loc), 'carbon')
    Fish = count_names(all_footprints(loc), 'fish')
    
    Names = [Cropland, Grazing, Forest, Carbon, Fish]
    Labels = show_labels(loc,["Cropland", "Grazing", "Forest", "Carbon", "Fish"])
    Explode = [0,0,0,0,0]
    
    plt.pie(Names, explode=Explode, labels=Labels, autopct='%.0f%%', shadow=False)

    plt.show()
    
    return None
    
@typecheck
def main(filename: str) -> None:
    """
    Reads the file from given filename, analyzes the data,
    returns the result
    """
    #return None #stub
    # Template from HtDAP, based on composition
    return show_pie_chart(read(filename))
    
start_testing()

#tests for is_bigger
expect(is_bigger(AF_CRF, AF_GRF), True)
expect(is_bigger(AF_FOF, AF_GRF), False)
expect(is_bigger(AN_CAF, AF_GRF), False)

#tests for same_amt
expect(same_amt(AF_CRF, AF_GRF), False)
expect(same_amt(AF_FOF, AF_GRF), False)
expect(same_amt(AN_CAF, AF_GRF), True)

#tests for same_name
expect(same_name(AF_CRF, AF_GRF), False)
expect(same_name(AF_FOF, AF_GRF), False)
expect(same_name(AN_GRF, AF_GRF), True)

#tests for is_biggest
expect(is_biggest(LOF0), [])
expect(is_biggest(LOF1), [AF_CRF])
expect(is_biggest(LOF2), [AN_CRF])
expect(is_biggest(LOF4), [AF_CRF])
expect(is_biggest(LOF5), [AF_GRF, AN_CAF])
expect(is_biggest(LOF8), [AF_CRF])
expect(is_biggest([Footprint("grazing", 0.2), Footprint("fish", 0.09), Footprint("carbon", 0.2), 
            Footprint("cropland", 0.3), Footprint("grazing", 0.15)]), [Footprint("cropland", 0.3)])

#tests for biggest_footprint
expect(biggest_footprint(AFGHANISTAN), [Footprint('cropland', 0.3)])
expect(biggest_footprint(ALBANIA), [Footprint('carbon', 0.87)])
expect(biggest_footprint(ALGERIA), [Footprint('carbon', 1.14)])
expect(biggest_footprint(ANGOLA), [Footprint('cropland', 0.33)])

#tests for all_footprint
expect(all_footprints(LOC1), [Footprint('cropland', 0.3)])
expect(all_footprints(LOC2), [Footprint('cropland', 0.3), Footprint('carbon', 0.87), 
                              Footprint('carbon', 1.14), Footprint('cropland', 0.33)])
#tests for count_names
expect(count_names(LOF0, "grazing"), 0)
expect(count_names(LOF1, "grazing"), 1)
expect(count_names(LOF4, "carbon"), 2)

#tests for show_labels
expect(show_labels(LOC0,["Cropland", "Grazing", "Forest", "Carbon", "Fish"]), 
       [' ', ' ', ' ', ' ', ' '])
expect(show_labels(LOC1,["Cropland", "Grazing", "Forest", "Carbon", "Fish"]), 
      ['Cropland', ' ', ' ', ' ', ' '])
expect(show_labels(LOC3,["Cropland", "Grazing", "Forest", "Carbon", "Fish"]), 
      ['Cropland', ' ', ' ', 'Carbon', ' '])

#tests for show_pie_chart

#For "show_pie_chart" and "main" tests, see attached drawings for what the tests should return. 

#This should return an empty pie chart with no labels and one percentage at 0%. 
expect(show_pie_chart(LOC0), None)
#This should return a pie chart with one wedge representing 100% of the chart, labeled "Cropland"
expect(show_pie_chart(LOC1), None)
#This should return a pie chart with two wedges, each of witch represents 50% of the chart. The first wedge
#shoud be labeled "Cropland" and the second "Carbon". 
expect(show_pie_chart(LOC3), None)

#tests for main

#the tests for main should be identical to the tests in "show_pie_chart", in the same order, because they 
#are testing the same data sets 

#This should return an empty pie chart with no labels and one percentage at 0%. 
expect(main('smallest_test.csv'), None)
#This should return a pie chart with one wedge representing 100% of the chart, labeled "Cropland"
expect(main('testfile_one_line.csv'), None)
#This should return a pie chart with two wedges, each of witch represents 50% of the chart. The first wedge
#shoud be labeled "Cropland" and the second "Carbon". 
expect(main('testfile_four_lines.csv'), None)

summary()
````
